```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import timedelta

# Load data
df_inc = pd.read_excel("Incidents.xlsx")
df_maint = pd.read_excel("CHWY_maintenace_data.xlsx")

# First, let's check what columns we have
print("=== COLUMN INSPECTION ===")
print("Incident columns:", df_inc.columns.tolist())
print("\nMaintenance columns:", df_maint.columns.tolist())

# Convert date columns to datetime
df_inc['Incident_Date'] = pd.to_datetime(df_inc['Incident Date'])
# The maintenance date column is 'Bas. start date' (with period and space)
df_maint['Maintenance_Date'] = pd.to_datetime(df_maint['Bas. start date'])

print("=== DATA OVERVIEW ===")
print(f"Total Incidents: {len(df_inc)}")
print(f"Total Maintenance Records: {len(df_maint)}")
print(f"\nIncident Date Range: {df_inc['Incident_Date'].min()} to {df_inc['Incident_Date'].max()}")
print(f"Maintenance Date Range: {df_maint['Maintenance_Date'].min()} to {df_maint['Maintenance_Date'].max()}")

# ============================================
# FEATURE ENGINEERING FOR CORRELATION ANALYSIS
# ============================================

# 1. Time-based aggregation (monthly)
df_inc['Year_Month'] = df_inc['Incident_Date'].dt.to_period('M')
df_maint['Year_Month'] = df_maint['Maintenance_Date'].dt.to_period('M')

# Count incidents and maintenance per month
monthly_incidents = df_inc.groupby('Year_Month').size().reset_index(name='Incident_Count')
monthly_maint = df_maint.groupby('Year_Month').size().reset_index(name='Maintenance_Count')

# Merge monthly data
monthly_data = pd.merge(monthly_incidents, monthly_maint, on='Year_Month', how='outer').fillna(0)
monthly_data = monthly_data.sort_values('Year_Month')

# 2. Site-based analysis
site_incidents = df_inc.groupby('Site').size().reset_index(name='Incident_Count')
# The maintenance site column is 'MaintPlant'
site_maint = df_maint.groupby('MaintPlant').size().reset_index(name='Maintenance_Count')

# Check unique sites in both datasets
print("\n=== SITE MATCHING CHECK ===")
print(f"Unique sites in Incidents: {df_inc['Site'].nunique()}")
print(f"Unique sites in Maintenance (MaintPlant): {df_maint['MaintPlant'].nunique()}")
common_sites = set(df_inc['Site'].unique()) & set(df_maint['MaintPlant'].unique())
print(f"Common sites: {len(common_sites)}")
if len(common_sites) < 10:
    print(f"Sample common sites: {list(common_sites)[:10]}")

# 3. Equipment-based analysis (if equipment info exists in incidents)
# Create features: maintenance activity before incident
def count_maintenance_before_incident(incident_date, site, days_before=30):
    """Count maintenance activities X days before an incident"""
    start_date = incident_date - timedelta(days=days_before)
    # Match by site using MaintPlant column
    maint_count = len(df_maint[
        (df_maint['Maintenance_Date'] >= start_date) & 
        (df_maint['Maintenance_Date'] < incident_date) &
        (df_maint['MaintPlant'] == site)
    ])
    return maint_count

# Add features to incident data
df_inc['Maint_7days_before'] = df_inc.apply(
    lambda row: count_maintenance_before_incident(row['Incident_Date'], row['Site'], 7), axis=1
)
df_inc['Maint_30days_before'] = df_inc.apply(
    lambda row: count_maintenance_before_incident(row['Incident_Date'], row['Site'], 30), axis=1
)
df_inc['Maint_90days_before'] = df_inc.apply(
    lambda row: count_maintenance_before_incident(row['Incident_Date'], row['Site'], 90), axis=1
)

# 4. Create binary target: Did incident happen? (1=Yes, 0=No)
df_inc['Incident_Occurred'] = 1

# ============================================
# CORRELATION MATRIX
# ============================================

print("\n=== CORRELATION ANALYSIS ===")

# Monthly correlation
monthly_corr = monthly_data[['Incident_Count', 'Maintenance_Count']].corr()
print("\n1. MONTHLY TIME-SERIES CORRELATION:")
print(monthly_corr)
print(f"\nCorrelation coefficient: {monthly_corr.iloc[0,1]:.3f}")

# Feature correlation with incidents
feature_cols = ['Maint_7days_before', 'Maint_30days_before', 'Maint_90days_before']
correlation_summary = df_inc[feature_cols + ['Incident_Occurred']].corr()['Incident_Occurred'].drop('Incident_Occurred')
print("\n2. MAINTENANCE TIMING vs INCIDENT OCCURRENCE:")
print(correlation_summary)

# ============================================
# VISUALIZATIONS
# ============================================

fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Plot 1: Monthly trends
ax1 = axes[0, 0]
monthly_data['Year_Month_str'] = monthly_data['Year_Month'].astype(str)
x_pos = range(len(monthly_data))
ax1.plot(x_pos, monthly_data['Incident_Count'], marker='o', label='Incidents', linewidth=2)
ax1.plot(x_pos, monthly_data['Maintenance_Count'], marker='s', label='Maintenance', linewidth=2)
ax1.set_xlabel('Month')
ax1.set_ylabel('Count')
ax1.set_title('Monthly Trends: Incidents vs Maintenance')
ax1.legend()
ax1.grid(True, alpha=0.3)
# Show fewer x-axis labels for readability
step = max(1, len(monthly_data) // 10)
ax1.set_xticks(x_pos[::step])
ax1.set_xticklabels(monthly_data['Year_Month_str'].iloc[::step], rotation=45, ha='right')

# Plot 2: Correlation heatmap
ax2 = axes[0, 1]
sns.heatmap(monthly_corr, annot=True, cmap='coolwarm', center=0, ax=ax2, vmin=-1, vmax=1)
ax2.set_title('Correlation Matrix: Monthly Data')

# Plot 3: Scatter plot - Monthly incidents vs maintenance
ax3 = axes[1, 0]
ax3.scatter(monthly_data['Maintenance_Count'], monthly_data['Incident_Count'], alpha=0.6, s=100)
ax3.set_xlabel('Monthly Maintenance Count')
ax3.set_ylabel('Monthly Incident Count')
ax3.set_title('Scatter: Monthly Maintenance vs Incidents')
ax3.grid(True, alpha=0.3)

# Add trend line
z = np.polyfit(monthly_data['Maintenance_Count'], monthly_data['Incident_Count'], 1)
p = np.poly1d(z)
ax3.plot(monthly_data['Maintenance_Count'], p(monthly_data['Maintenance_Count']), 
         "r--", alpha=0.8, linewidth=2, label=f'Trend line')
ax3.legend()

# Plot 4: Maintenance before incidents
ax4 = axes[1, 1]
maint_before_data = df_inc[['Maint_7days_before', 'Maint_30days_before', 'Maint_90days_before']].mean()
ax4.bar(['7 days', '30 days', '90 days'], maint_before_data.values, color=['#FF6B6B', '#4ECDC4', '#45B7D1'])
ax4.set_ylabel('Avg Maintenance Activities')
ax4.set_title('Average Maintenance Before Incidents')
ax4.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('maintenance_incident_correlation.png', dpi=300, bbox_inches='tight')
plt.show()

# ============================================
# STATISTICAL INSIGHTS
# ============================================

print("\n=== KEY INSIGHTS ===")
print(f"\n1. Average maintenance activities 30 days before incident: {df_inc['Maint_30days_before'].mean():.2f}")
print(f"   Standard deviation: {df_inc['Maint_30days_before'].std():.2f}")

# Incidents with/without recent maintenance
incidents_with_maint = len(df_inc[df_inc['Maint_30days_before'] > 0])
incidents_without_maint = len(df_inc[df_inc['Maint_30days_before'] == 0])
print(f"\n2. Incidents with maintenance in prior 30 days: {incidents_with_maint} ({incidents_with_maint/len(df_inc)*100:.1f}%)")
print(f"   Incidents without maintenance in prior 30 days: {incidents_without_maint} ({incidents_without_maint/len(df_inc)*100:.1f}%)")

# Site comparison
print("\n3. Top 5 sites by incident count:")
top_sites = df_inc['Site'].value_counts().head()
print(top_sites)

# ============================================
# PREPARE DATA FOR ML MODELING
# ============================================

print("\n=== PREPARING DATA FOR ML ===")

# Create a comprehensive feature set
ml_features = df_inc[[
    'Site', 'Department', 'Type',
    'Maint_7days_before', 'Maint_30days_before', 'Maint_90days_before'
]].copy()

# Add time features
ml_features['Month'] = df_inc['Incident_Date'].dt.month
ml_features['Quarter'] = df_inc['Incident_Date'].dt.quarter
ml_features['DayOfWeek'] = df_inc['Incident_Date'].dt.dayofweek

# Target variable
ml_features['Target'] = df_inc['Incident_Occurred']

print(f"\nML Dataset shape: {ml_features.shape}")
print(f"Features: {ml_features.columns.tolist()}")
print(f"\nSample of features:")
print(ml_features.head())

# Check for missing values
print(f"\nMissing values:")
print(ml_features.isnull().sum())

# Save prepared data
ml_features.to_csv('ml_ready_data.csv', index=False)
print("\n ML-ready dataset saved as 'ml_ready_data.csv'")

print("\n=== NEXT STEPS FOR ML MODEL ===")
print("""
1. Encode categorical variables (Site, Department, Type)
2. Handle class imbalance if needed
3. Split data into train/test sets
4. Try models: Logistic Regression, Random Forest, XGBoost
5. Evaluate with: Accuracy, Precision, Recall, F1-Score, AUC-ROC
""")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import timedelta

# Load data
df_inc = pd.read_excel("Incidents.xlsx")
df_maint = pd.read_excel("CHWY_maintenace_data.xlsx")

# Convert date columns to datetime
df_inc['Incident_Date'] = pd.to_datetime(df_inc['Incident Date'])
df_maint['Maintenance_Date'] = pd.to_datetime(df_maint['Bas. start date'])

print("=== DATA OVERVIEW ===")
print(f"Total Incidents: {len(df_inc)}")
print(f"Total Maintenance Records: {len(df_maint)}")
print(f"\nIncident Date Range: {df_inc['Incident_Date'].min()} to {df_inc['Incident_Date'].max()}")
print(f"Maintenance Date Range: {df_maint['Maintenance_Date'].min()} to {df_maint['Maintenance_Date'].max()}")

# ============================================
# ANALYZE: MAINTENANCE BEFORE EACH INCIDENT
# ============================================

print("\n" + "="*70)
print("ANALYZING: WAS THERE MAINTENANCE BEFORE EACH INCIDENT?")
print("="*70)

def count_maintenance_in_window(incident_date, days_before, days_after=0):
    """Count maintenance activities in a time window around an incident"""
    start_date = incident_date - timedelta(days=days_before)
    end_date = incident_date + timedelta(days=days_after)
    
    count = len(df_maint[
        (df_maint['Maintenance_Date'] >= start_date) & 
        (df_maint['Maintenance_Date'] <= end_date)
    ])
    return count

def had_maintenance_in_window(incident_date, days_before, days_after=0):
    """Check if ANY maintenance happened in time window"""
    return count_maintenance_in_window(incident_date, days_before, days_after) > 0

# Add features for different time windows
print("\nCalculating maintenance activity around each incident...")

time_windows = [
    (1, "1 day before"),
    (3, "3 days before"),
    (7, "7 days before"),
    (14, "14 days before"),
    (30, "30 days before"),
    (60, "60 days before"),
    (90, "90 days before")
]

for days, label in time_windows:
    col_count = f'Maint_Count_{days}d_before'
    col_binary = f'Had_Maint_{days}d_before'
    
    df_inc[col_count] = df_inc['Incident_Date'].apply(
        lambda date: count_maintenance_in_window(date, days, 0)
    )
    df_inc[col_binary] = df_inc[col_count] > 0

# ============================================
# ANALYZE: MAINTENANCE AFTER EACH INCIDENT
# ============================================

print("\n" + "="*70)
print("ANALYZING: WAS THERE MAINTENANCE AFTER EACH INCIDENT?")
print("="*70)

time_windows_after = [
    (1, "1 day after"),
    (3, "3 days after"),
    (7, "7 days after"),
    (14, "14 days after"),
    (30, "30 days after"),
    (60, "60 days after"),
    (90, "90 days after")
]

for days, label in time_windows_after:
    col_count = f'Maint_Count_{days}d_after'
    col_binary = f'Had_Maint_{days}d_after'
    
    df_inc[col_count] = df_inc['Incident_Date'].apply(
        lambda date: count_maintenance_in_window(date, 0, days)
    )
    df_inc[col_binary] = df_inc[col_count] > 0

# ============================================
# SUMMARY STATISTICS
# ============================================

print("\n" + "="*70)
print("SUMMARY: MAINTENANCE BEFORE INCIDENTS")
print("="*70)

for days, label in time_windows:
    col_binary = f'Had_Maint_{days}d_before'
    col_count = f'Maint_Count_{days}d_before'
    
    pct_with_maint = (df_inc[col_binary].sum() / len(df_inc)) * 100
    avg_maint_count = df_inc[col_count].mean()
    max_maint_count = df_inc[col_count].max()
    
    print(f"\n{label.upper()}:")
    print(f"   {df_inc[col_binary].sum():,} incidents ({pct_with_maint:.1f}%) had maintenance")
    print(f"   {len(df_inc) - df_inc[col_binary].sum():,} incidents ({100-pct_with_maint:.1f}%) had NO maintenance")
    print(f"   Average maintenance count: {avg_maint_count:.2f}")
    print(f"   Maximum maintenance count: {max_maint_count:.0f}")

print("\n" + "="*70)
print("SUMMARY: MAINTENANCE AFTER INCIDENTS")
print("="*70)

for days, label in time_windows_after:
    col_binary = f'Had_Maint_{days}d_after'
    col_count = f'Maint_Count_{days}d_after'
    
    pct_with_maint = (df_inc[col_binary].sum() / len(df_inc)) * 100
    avg_maint_count = df_inc[col_count].mean()
    max_maint_count = df_inc[col_count].max()
    
    print(f"\n{label.upper()}:")
    print(f"   {df_inc[col_binary].sum():,} incidents ({pct_with_maint:.1f}%) had maintenance after")
    print(f"   {len(df_inc) - df_inc[col_binary].sum():,} incidents ({100-pct_with_maint:.1f}%) had NO maintenance after")
    print(f"   Average maintenance count: {avg_maint_count:.2f}")
    print(f"   Maximum maintenance count: {max_maint_count:.0f}")

# ============================================
# CORRELATION ANALYSIS
# ============================================

print("\n" + "="*70)
print("CORRELATION: MAINTENANCE TIMING vs INCIDENT OCCURRENCE")
print("="*70)

# Create correlation matrix for before/after maintenance
maint_cols_before = [f'Maint_Count_{days}d_before' for days, _ in time_windows]
maint_cols_after = [f'Maint_Count_{days}d_after' for days, _ in time_windows_after]

# Correlation with incident occurrence (all incidents have occurred, so correlation with other factors)
correlation_data = df_inc[maint_cols_before + maint_cols_after].copy()

# Calculate correlations between different time windows
corr_matrix = correlation_data.corr()

print("\nCorrelation Matrix Preview (First few columns):")
print(corr_matrix.iloc[:5, :5].round(3))

# ============================================
# KEY INSIGHT: COMPARE BEFORE VS AFTER
# ============================================

print("\n" + "="*70)
print("KEY INSIGHT: PREVENTIVE vs REACTIVE MAINTENANCE")
print("="*70)

# 30-day window comparison
before_30 = df_inc['Had_Maint_30d_before'].sum()
after_30 = df_inc['Had_Maint_30d_after'].sum()

print(f"\n30-DAY WINDOW COMPARISON:")
print(f"   Incidents with maintenance BEFORE: {before_30:,} ({before_30/len(df_inc)*100:.1f}%)")
print(f"   Incidents with maintenance AFTER: {after_30:,} ({after_30/len(df_inc)*100:.1f}%)")

if after_30 > before_30:
    print(f"\n   REACTIVE: {after_30-before_30:,} more incidents had maintenance AFTER")
    print("      Suggests maintenance is often a response to incidents")
elif before_30 > after_30:
    print(f"\n   PREVENTIVE: {before_30-after_30:,} more incidents had maintenance BEFORE")
    print("      Suggests maintenance may not be preventing incidents effectively")
else:
    print("\n   BALANCED: Equal maintenance before and after incidents")

# ============================================
# VISUALIZATIONS
# ============================================

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('Maintenance Before/After Incidents Analysis', fontsize=16, fontweight='bold', y=1.00)

# Plot 1: Percentage of incidents with maintenance BEFORE
ax1 = axes[0, 0]
before_pcts = [
    (df_inc[f'Had_Maint_{days}d_before'].sum() / len(df_inc)) * 100 
    for days, _ in time_windows
]
days_labels = [f'{days}d' for days, _ in time_windows]
bars1 = ax1.bar(days_labels, before_pcts, color='#4ECDC4', edgecolor='black', linewidth=1.5)
ax1.set_ylabel('% of Incidents', fontsize=11)
ax1.set_xlabel('Days Before Incident', fontsize=11)
ax1.set_title('Incidents with Maintenance BEFORE', fontsize=12, fontweight='bold')
ax1.grid(True, alpha=0.3, axis='y')
for bar in bars1:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.1f}%', ha='center', va='bottom', fontsize=9)

# Plot 2: Percentage of incidents with maintenance AFTER
ax2 = axes[0, 1]
after_pcts = [
    (df_inc[f'Had_Maint_{days}d_after'].sum() / len(df_inc)) * 100 
    for days, _ in time_windows_after
]
bars2 = ax2.bar(days_labels, after_pcts, color='#FF6B6B', edgecolor='black', linewidth=1.5)
ax2.set_ylabel('% of Incidents', fontsize=11)
ax2.set_xlabel('Days After Incident', fontsize=11)
ax2.set_title('Incidents with Maintenance AFTER', fontsize=12, fontweight='bold')
ax2.grid(True, alpha=0.3, axis='y')
for bar in bars2:
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.1f}%', ha='center', va='bottom', fontsize=9)

# Plot 3: Before vs After comparison
ax3 = axes[0, 2]
x = np.arange(len(days_labels))
width = 0.35
bars3a = ax3.bar(x - width/2, before_pcts, width, label='Before', color='#4ECDC4', edgecolor='black')
bars3b = ax3.bar(x + width/2, after_pcts, width, label='After', color='#FF6B6B', edgecolor='black')
ax3.set_ylabel('% of Incidents', fontsize=11)
ax3.set_xlabel('Time Window (days)', fontsize=11)
ax3.set_title('BEFORE vs AFTER Comparison', fontsize=12, fontweight='bold')
ax3.set_xticks(x)
ax3.set_xticklabels(days_labels)
ax3.legend()
ax3.grid(True, alpha=0.3, axis='y')

# Plot 4: Average maintenance count BEFORE
ax4 = axes[1, 0]
before_avg = [df_inc[f'Maint_Count_{days}d_before'].mean() for days, _ in time_windows]
ax4.plot(days_labels, before_avg, marker='o', linewidth=3, markersize=10, color='#4ECDC4')
ax4.set_ylabel('Avg Maintenance Count', fontsize=11)
ax4.set_xlabel('Days Before Incident', fontsize=11)
ax4.set_title('Average Maintenance Count BEFORE', fontsize=12, fontweight='bold')
ax4.grid(True, alpha=0.3)
ax4.fill_between(range(len(days_labels)), before_avg, alpha=0.3, color='#4ECDC4')

# Plot 5: Average maintenance count AFTER
ax5 = axes[1, 1]
after_avg = [df_inc[f'Maint_Count_{days}d_after'].mean() for days, _ in time_windows_after]
ax5.plot(days_labels, after_avg, marker='s', linewidth=3, markersize=10, color='#FF6B6B')
ax5.set_ylabel('Avg Maintenance Count', fontsize=11)
ax5.set_xlabel('Days After Incident', fontsize=11)
ax5.set_title('Average Maintenance Count AFTER', fontsize=12, fontweight='bold')
ax5.grid(True, alpha=0.3)
ax5.fill_between(range(len(days_labels)), after_avg, alpha=0.3, color='#FF6B6B')

# Plot 6: Correlation heatmap (subset)
ax6 = axes[1, 2]
corr_subset = corr_matrix.iloc[:4, :4]
sns.heatmap(corr_subset, annot=True, fmt='.2f', cmap='coolwarm', center=0, 
            ax=ax6, vmin=-1, vmax=1, cbar_kws={'label': 'Correlation'})
ax6.set_title('Correlation Matrix (Sample)', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.savefig('maintenance_before_after_incidents.png', dpi=300, bbox_inches='tight')
print("\n Visualization saved as 'maintenance_before_after_incidents.png'")
plt.show()

# ============================================
# EXPORT RESULTS
# ============================================

# Save detailed results
results_df = df_inc[[
    'Incident No.', 'Incident_Date', 'Site', 'Type', 'Brief Description'
] + maint_cols_before + maint_cols_after].copy()

results_df.to_csv('incidents_with_maintenance_timing.csv', index=False)
print(" Detailed results saved as 'incidents_with_maintenance_timing.csv'")

# Create summary table
summary_data = []
for days, label in time_windows:
    col_binary = f'Had_Maint_{days}d_before'
    col_count = f'Maint_Count_{days}d_before'
    summary_data.append({
        'Time_Window': label,
        'Direction': 'Before',
        'Incidents_With_Maintenance': df_inc[col_binary].sum(),
        'Percentage': (df_inc[col_binary].sum() / len(df_inc)) * 100,
        'Avg_Maintenance_Count': df_inc[col_count].mean()
    })

for days, label in time_windows_after:
    col_binary = f'Had_Maint_{days}d_after'
    col_count = f'Maint_Count_{days}d_after'
    summary_data.append({
        'Time_Window': label,
        'Direction': 'After',
        'Incidents_With_Maintenance': df_inc[col_binary].sum(),
        'Percentage': (df_inc[col_binary].sum() / len(df_inc)) * 100,
        'Avg_Maintenance_Count': df_inc[col_count].mean()
    })

summary_df = pd.DataFrame(summary_data)
summary_df.to_csv('maintenance_timing_summary.csv', index=False)
print(" Summary table saved as 'maintenance_timing_summary.csv'")

print("\n" + "="*70)
print("ANALYSIS COMPLETE!")
print("="*70)
print("\nFiles created:")
print("  1. maintenance_before_after_incidents.png - Visualizations")
print("  2. incidents_with_maintenance_timing.csv - Detailed data for each incident")
print("  3. maintenance_timing_summary.csv - Summary statistics")
print("\nUse this data for ML modeling to predict incident likelihood based on maintenance patterns!")
```
